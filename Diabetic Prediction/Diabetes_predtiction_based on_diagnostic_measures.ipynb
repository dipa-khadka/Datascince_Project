# **About Dataset :-**
The datasets consist of several medical predictor (independent) variables and one target (dependent) variable, Outcome. Independent variables include the number of pregnancies the patient has had, their BMI, insulin level, age, and so on.

* Pregnancies : The number of times a woman has been pregnant. 
* Glucose :  The concentration of glucose (a type of sugar) in a person's blood.
* BloodPressure :  The measurement of a person's blood pressure. 
* SkinThickness : The thickness of the skinfold at a specific site on the body, often measured in millimeters.
* Insulin : The concentration or level of insulin in a person's blood. 
* BMI : Body Mass Index (Indicator of body fatness and overall health).
* DiabetesPedigreeFunction : numerical score that represents The likelihood of diabetes based on family history. 
* Age :  The age of an individual at the time of data collection or measurement.
* Outcome : whether an individual has diabetes or not.

```python
X = df.drop(columns='Outcome',axis=1)
Y = df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.2, random_state=123)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X_train, Y_train)
```

```python
logreg_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, logreg_pred)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
print(confusion_matrix(Y_test, logreg_pred))
```

```python
print(classification_report(Y_test, logreg_pred))
```

```python
from sklearn.svm import SVC
from sklearn import svm
```

```python
svm_model = svm.SVC(kernel='linear')
```

```python
svm_model.fit(X_train, Y_train)
```

```python
svm_pred = svm_model.predict(X_test)
print(accuracy_score(Y_test, svm_pred))
print(confusion_matrix(Y_test, svm_pred))
print(classification_report(Y_test, svm_pred))
```
